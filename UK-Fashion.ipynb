RECENCY CALCULATE

```python
Last_date_purchased = Merged_Items.groupby('CustomerID', as_index=False).agg(Last_date_purchased = ('TransactionDate','max'))

Last_date_purchased
```

```python
Last_date_purchased['Last_date_purchased'] = pd.to_datetime(Last_date_purchased['Last_date_purchased']).dt.date
```

```python
Last_date_purchased['Count_Date_Recency'] = Last_date_purchased['Last_date_purchased'].apply(lambda x: (Report_date-x).days)

Last_date_purchased
```

Frequency Calculate

```python
Count_Frequency = Merged_Items.groupby('CustomerID',as_index=False).agg(Count_Frequency = ('TransactionDate','count'))

Count_Frequency
```

Monetary Calculation

```python
Count_Monetary = Merged_Items.groupby('CustomerID',as_index=False).agg(Count_Monetary = ('Revenue','sum'))

Count_Monetary
```

Combine RFM

```python
RFM = pd.merge(Last_date_purchased,Count_Frequency, on='CustomerID', how='outer')
```

```python
RFM = pd.merge(RFM,Count_Monetary, on='CustomerID', how='outer')
```

```python
RFM
```

Describe to get N-title

```python
RFM.describe()
```

```python
RFM['r_quartile'] = pd.qcut(RFM['Count_Date_Recency'], 4, ['1','2','3','4'])
RFM['f_quartile'] = pd.qcut(RFM['Count_Frequency'], 4, ['4','3','2','1'])
RFM['m_quartile'] = pd.qcut(RFM['Count_Monetary'], 4, ['4','3','2','1'])
```

```python
RFM
```

```python
RFM['RFM_Score'] = RFM['r_quartile'].astype(str) + RFM['f_quartile'].astype(str) + RFM['m_quartile'].astype(str)

RFM
```

```python
RFM.groupby('RFM_Score',as_index=False).agg(cat_count=('RFM_Score','count'))
```

```python
def RFM_Segmentation(RFM_Score):
    if RFM_Score in ['444','443','434','433','344','343','334']:
        return 'VVIP'
    elif RFM_Score in ['423','333','323']:
        return 'VIP'
    elif RFM_Score in ['432','422','332']:
        return 'New Potential'
    elif RFM_Score in ['421','411','322','321','312','311']:
        return 'New Customer'
    elif RFM_Score in ['244','243','234','233','144','143','134','133','123']:
        return 'VIP_Dont_Lost'
    else:
        return 'Lost'
```

```python
RFM['RFM_Segmentation'] = RFM['RFM_Score'].apply(RFM_Segmentation)

RFM
```
